```python
import pandas as pd
import matplotlib.pyplot as plt
cars=pd.read_excel('cars_ds.xlsx')
cars
```

```python
cars['Avg_mileage']=(cars['City_Mileage_km_litre']+cars['Highway_Mileage_km_litre'])/2
cars
```

```python
# TASK 1
cars1=cars.groupby(by=['Make'])['City_Mileage_km_litre'].mean()
cars1.sort_values(ascending=False).head(10)
```

```python
# TASK 2
cars1.sort_values(ascending= False)[10:20]
```

```python
# TASK 3
cars2=cars.groupby(by=['Body_Type'])['isker'].count()
cars2
```

```python
cars2.plot(x='Body_Type',kind='pie',autopct='%f.2 %%',figsize=(12,12))
```

```python
print('The body type SUV is most popular in the given market')
```

```python
# TASK 4
cars4=cars[(cars.Make =='Hyundai')|(cars.Make =='Mahindra')|(cars.Make =='Renault')|(cars.Make =='Skoda')]
cars41=cars4.groupby(by='Make')['City_Mileage_km_litre','Highway_Mileage_km_litre'].mean()
cars41
```

```python
cars42=pd.DataFrame({'Make':['Hyundai','Mahindra','Renault','Skoda'],'City Mileage':[16.67,20.00,20.55,13.95],'Highway Mileage':[18.29,18.14,17.69,17.26]})
cars42
```

```python
cars42.plot(x='Make',kind='bar',figsize=(10,5))
```

```python
print('As we can see here Renault is offering highest City Mileage whereas Hyundai is offering highest Highway Mileage')
```
