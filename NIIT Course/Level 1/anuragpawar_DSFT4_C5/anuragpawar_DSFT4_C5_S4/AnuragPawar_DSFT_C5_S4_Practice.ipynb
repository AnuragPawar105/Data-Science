```python
# TASK 1
X=[6,7,8,9,10,11,12,13,14]
p_x=[0.03,0.08,0.15,0.20,0.19,0.16,0.10,0.07,0.02]
```

```python
# TASK 1b
#P2=p(12 days)-p(8 days)
P_12=0.03+0.08+0.15+0.20+0.19+0.16+0.10
P_8=0.03+0.08
P2=P_12-P_8
print('The probability of 8-12 days lost is',P2)
```

```python
# TASK 1d
import math
mean=0
for x,p in zip(X,p_x):
    mean=mean+x*p
print('Mean of x is %1.4f'%mean)
var=0
for x,p in zip(X,p_x):
    var=var+((x-mean)**2)*p
sd=math.sqrt(var)
print('Standard deviation is %1.4f'%sd)
```

```python
# TASK 1e
import matplotlib.pyplot as plt
plt.plot(X,p_x,'o-')
plt.title('Probability Distribution')
plt.xlabel('Number of days lost')
plt.ylabel('Probability')
```

```python
# TASK 4
import pandas as pd
import scipy.stats as sts
w1=[3,6,4,6,2,3,1,5,1,0,3,3]
w2=[1,2,4,0,2,6,5,4,2,5,3,4]
w3=[5,3,5,3,5,4,7,3,4,8,3,1]
w=pd.DataFrame([w1,w2,w3],index=['week1','week2','week3'])
rate=w.mean().mean()
n2=np.arange(0,31)
cus_pmf=sts.poisson.pmf(n2, rate)
cus_pmf
```

```python
# TASK 4a
print('The probability that no customers arrive in a five-minute interval of time is %1.3f'%cus_pmf[0])
```

```python
# TASK 4b
print('The probability that six or more customers arrive in a 10-minute interval of time is %1.3f'%(1-(np.sum(cus_pmf[0:6]))))
```
