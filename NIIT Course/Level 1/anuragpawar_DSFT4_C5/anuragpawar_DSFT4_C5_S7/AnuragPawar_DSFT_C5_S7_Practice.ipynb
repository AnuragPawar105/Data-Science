# TASK 1

Null Hypothesis: Mean Score > 71

Alternate Hypothesis: Mean Score<= 71

```python
import statistics as st
sample_data=[61,93,75,67,84,95]
s_mean=st.mean(sample_data)
s_sd=st.stdev(sample_data)
n=len(sample_data)
p_mean=71
alpha= 0.05
dof=n-1
```

```python
import math
from scipy.stats import t

t_statistics= (s_mean - p_mean)/(s_sd/math.sqrt(n))
print('The t-statistics is ',t_statistics)

t_critical= t.ppf(alpha, dof)
print('The t-critical value is ',t_critical)
```

As per the decision rule:

t-statistics= 1.44 > -t-critical= -2.015, which leads to the acceptance of the null hypothesis.

```python
# Plotting the curve
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
l=s_mean-(4*s_sd)
u=s_mean+(4*s_sd)
r=np.arange(-4, 4, 0.1)
plt.plot(r, sts.t.pdf(r, 5))
plt.title("Distribution Curve")
plt.xlabel("Marks Obtained")
plt.ylabel("Probability Density")
plt.show()
```

# Conclusion

Since the null hypothesis is accepted, Oscar can have confidence that the mean score for the class on the test would be above 71.

# TASK 3

Null Hypothesis: The average fill for all four machines is same.

Alternate Hypothesis: The average fill for all four machines is not same.

```python
# Accessing the data
mach1=[4.05, 4.01, 4.04, 4.02]
mach2=[4.02, 3.99, 4.01, 3.99, 4.00, 4.00]
mach3=[4.00, 3.97, 3.95, 3.98, 3.97]
mach4=[4.01, 4.00, 3.99, 4.02]
```

```python
# Calculating total fill for each machine
T1=sum(mach1)
T2=sum(mach2)
T3=sum(mach3)
T4=sum(mach4)
T=T1+T2+T3+T4
print('Total fill for machine 1 =',T1)
print('Total fill for machine 2 =',T2)
print('Total fill for machine 3 =',T3)
print('Total fill for machine 4 =',T4)
print('Total fill for all four machines =',T)
```

```python
# Sample size
n1=len(mach1)
n2=len(mach2)
n3=len(mach3)
n4=len(mach4)
n=n1+n2+n3+n4
print('Size of machine 1 =',n1)
print('Size of machine 2 =',n2)
print('Size of machine 3 =',n3)
print('Size of machine 4 =',n4)
print('Size of all machine =',n)
```

```python
# Calculating the mean
x1=st.mean(mach1)
x2=st.mean(mach2)
x3=st.mean(mach3)
x4=st.mean(mach4)
x=(x1+x2+x3+x4)/4
print('Mean of the machine 1 =',x1)
print('Mean of the machine 2 =',x2)
print('Mean of the machine 3 =',x3)
print('Mean of the machine 4 =',x4)
print('Mean of all the machines =',x)
```

```python
# calculating the sum of squares columns(SSC)
SSC=n1*(x1-x)**2 + n2*(x2-x)**2 + n3*(x3-x)**2 + n4*(x4-x)**2
print('SSC =',SSC)
```

```python
#calculating the sum of squares errors(SSE)
SSE=0
for i in mach1:
    SSE=SSE+(i-x1)**2
for i in mach2:
    SSE=SSE+(i-x2)**2
for i in mach3:
    SSE=SSE+(i-x3)**2
for i in mach4:
    SSE=SSE+(i-x4)**2
    
print('SSE =',SSE)
```

```python
# calculating the total sum of squares(SST)
SST=0
for i in mach1:
    SST=SST+(i-x)**2
for i in mach2:
    SST=SST+(i-x)**2
for i in mach3:
    SST=SST+(i-x)**2
for i in mach4:
    SST=SST+(i-x)**2
    
print('SST =',SST)
```

```python
# Calculating column, error and the total degree of freedom (dfc,dfe,dft)
dfc= 4-1
dfe= 19-4
dft= 19-1
print('dfc =',dfc)
print('dfe =',dfe)
print('dft =',dft)
```

```python
# calculating MSC
MSC= SSC/dfc
print('MSC =',MSC)
```

```python
# calculating MSE
MSE=SSE/dfe
print('MSE =',MSE)
```

```python
# calculating F-statistic
f_statistic=MSC/MSE
print('F-statistic =',f_statistic)
```

```python
# calculating F-critical
from scipy.stats import f
f_critical=f.ppf(1-0.01, dfc, dfe)
print('F-critical =',f_critical)
```

As per the decision rule:

F-statistic= 10.16 > F-critical=5.41, which leada to the rejection of null hypothesis.

```python
# plotting f-distribution
rv = f(dfc, dfe, loc=0, scale=1)
x= np.linspace(rv.ppf(0.01), rv.ppf(1-0.01), 100)
y= rv.pdf(x)

plt.xlim(0,7)
plt.plot(x,y,'r-')
plt.show()
```

# Conclusion

With the rejection of null hypothesis, we can conclude that there is a significant difference in the fill averages of all the four machines.
