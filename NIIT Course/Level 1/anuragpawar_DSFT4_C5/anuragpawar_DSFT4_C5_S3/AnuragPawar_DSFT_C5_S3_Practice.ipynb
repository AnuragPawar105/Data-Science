P(A)= Probability that the worker is female =0.46

P(B)= Probability that the worker is part-time =0.174

P(A n B)= Probability that the worker is female part-time worker =0.46* 0.25 =0.115

P(A|B)= Probability that the worker is female if a part-time worker is randomly selected

P(A|B)= P(A n B)/P(B)= 0.115/0.174= 0.66

```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics as st
NPD=pd.read_csv('DS1_C5_S3_Sampling_Practice_GamesSales_Data.csv')
NPD
```

```python
# Task 3.1
# 3.1.1
s_data1=NPD.sample(n=30, replace=False, random_state=0)
s_data1
```

```python
#3.1.2
s_data2=NPD.sample(n=30, replace= True, random_state=0)
s_data2
```

```python
#3.1.3
import numpy as np
index=np.arange(0, len(NPD), step=10)
s_data3=NPD.iloc[index]
s_data3
```

```python
#3.1.4
gr_data1=NPD.groupby(by=['Year'])['Genre'].value_counts()
gr_data2=NPD.groupby(['Year','Genre'])
s_data4=pd.DataFrame()

for i in gr_data1.index:
    data=gr_data2.get_group(i)
    sample_data=data.sample(n=5, replace=True, random_state= 1)
    s_data4=s_data4.append(sample_data, ignore_index= True)
    
s_data4
```

```python
# TASK 3.1.5
gr_data3=NPD.groupby('Year')
year=[2015,2020]
s_data5=pd.DataFrame()

for y in year:
    data1=gr_data3.get_group(y)
    sample_data1=data1.sample(n=50, replace=True, random_state=1)
    s_data5=s_data5.append(sample_data1, ignore_index=True)
s_data5
```

```python
gr_data4=s_data5.groupby('Genre')
genre=['Sports','Action']
s_data6=pd.DataFrame()

for g in genre:
    data2=gr_data4.get_group(g)
    sample_data2=data2.sample(n=16, replace=True, random_state=1)
    s_data6=s_data6.append(sample_data2, ignore_index= True)
s_data6
```
