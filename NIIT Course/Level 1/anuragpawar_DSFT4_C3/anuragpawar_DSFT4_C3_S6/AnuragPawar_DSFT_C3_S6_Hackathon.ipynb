```python
#     THE SHOPPING CART PROGRAM
Qavailable=[5,10,20,20,30]
itemcost=[20.50,90.00,200.00,39.99,79.99]
item=['Biscuits','Cereals','Chicken','Oats','Rice']
#the menu
print('                         ESSENTIAL ITEMS AVAILABLE')
print('              S.No              Items     Quantity    Cost/Item')
print('                 1           Biscuits         ',Qavailable[0],'      ',itemcost[0])
print('                 2            Cereals        ',Qavailable[1],'      ',itemcost[0])
print('                 3            Chicken        ',Qavailable[2],'     ',itemcost[2])
print('                 4               Oats        ',Qavailable[3],'      ',itemcost[3])
print('                 5               Rice        ',Qavailable[4],'      ',itemcost[4])
O0=int(input('How many biscuits packets are required ? '))
O1=int(input('How many cereal packets are required ? '))
O2=int(input('How many frozen chicken packets are required ? '))
O3=int(input('How many oats packets are required ? '))
O4=int(input('How many rice packets are required ? '))
print(' ')
if O0 > Qavailable[0]:
    print(item[0],' Out of stock')
if O1 > Qavailable[1]:
    print(item[1],' Out of stock')
if O2 > Qavailable[2]:
    print(item[2],' Out of stock')
if O3 > Qavailable[3]:
    print(item[3],' Out of stock')
if O4 > Qavailable[4]:
    print(item[4],' Out of stock')

print('FILL IN YOUR DELIVERY DETAILS')
name=input('Enter your name: ')
add=input('Enter your address: ')
print('Distance from store 5/10/15: 10km')
DC=30
ordervalue=[itemcost[0]*O0,itemcost[1]*O1,itemcost[2]*O2,itemcost[3]*O3,itemcost[4]*O4]
print('Delivery Charge: ',30,' will be levied for 10km of distance')
#the bill
print('                                    BILL')
print('              S.No          Item    Quantity   TotalCost')
print('                 1      ',item[0],'   ',O0,'      ',ordervalue[0])
print('                 2      ',item[1],'    ',O1,'      ',ordervalue[1])
print('                 3      ',item[2],'    ',O2,'      ',ordervalue[2])
print('                 4      ',item[3],'       ',O3,'      ',ordervalue[3])
print('                 5      ',item[4],'       ',O4,'      ',ordervalue[4])
from functools import reduce
Total_cost=reduce(lambda a,b:a+b, ordervalue)
Total_bill= Total_cost + DC
print(' ')
print('Total Item Cost: ',Total_cost)
print(' ')
print('Total Bill Amount: Total Item Cost + Delivery Charges: ',Total_bill)
print(' ')
print('Name: ',name)
print('Address:',add)
print('Have a nice day!!')
```
