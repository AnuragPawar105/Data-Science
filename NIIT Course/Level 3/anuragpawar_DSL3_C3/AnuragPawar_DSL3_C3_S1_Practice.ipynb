**Task 2**

```python
class Crawl(Crawler):
  def visit(self, link, source=None):
    st = 'visited : ' + str(link.url) + '| FROM : ' + str(link.referrer)
    print(st)
  def fail(self, link):
    print('Failed : ',repr(link.url))
  
url = 'https://www.nature.com/'
p = Crawl(links=[url], delay=0.02)
num_visited = 0
while (not  p.done and num_visited < 20):
  p.crawl(method=DEPTH, cahed=False, throttle=1)
  num_visited = num_visited + 1
```

```python
dic = p.visited
n = 10
show_10 = dict(list(dic.items())[0:n])
print('Showing the Visited sites :\n ')
for i in show_10:
  print(i)
```

```python
# Extracting the links

url = 'https://www.nature.com'

response = requests.get(url)

if response.status_code == 200:
  print('Connection sucessful : ',response.status_code)
else:
  print('Site is not recheable')

soup = BeautifulSoup(response.text, 'lxml')     # text = can write 'response.content' as well both are same
print(soup.prettify ) 
```

```python
# Extracting the link

tags = soup.find_all('a')
for tag in tags:
  print(tag.get('href'))
```

```python
# Extracting the paragraph

para = soup.find_all('p')
for i in para:
    print(i.text)
```

```python
# Extracting the Title

tags = soup.find_all('title')
for tag in tags:
  print(tag.get_text())
```
