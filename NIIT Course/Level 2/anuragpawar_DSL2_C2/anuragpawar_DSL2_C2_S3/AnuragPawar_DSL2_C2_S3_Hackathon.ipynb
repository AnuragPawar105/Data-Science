```python
import requests
from bs4 import BeautifulSoup
```

```python
#For 2015 
url= "https://en.wikipedia.org/wiki/List_of_horror_films_of_2015"
response= requests.get(url).text
soup= BeautifulSoup(response, "html")
print(soup.prettify)
```

```python
table= soup.find_all('table')
table1=table[1]
table1
```

```python
dir1=[]
for row in table1.find_all('tr')[1:]:
    for columns in row.find_all('td')[0:1]:
        dir1.append(columns.text.replace("\n",""))
dir1.pop(4)
dir1
```

```python
con1=[]
for row in table1.find_all('tr')[1:]:
    for columns in row.find_all('td')[-2:-1]:
        con1.append(columns.text.replace("\n",""))
con1.pop(4)
con1
```

```python
cast1=[]
for row in table1.find_all('tr')[1:]:
    for columns in row.find_all('td')[1:2]:
        cast1.append(columns.text.replace("\n",""))
cast1.pop(4)
cast1
```

```python
title1=[]
for row in table1.find_all('th', scope='row')[:]:
    title1.append(row.text.replace("\n",""))
title1
```

```python
import pandas as pd
import numpy as np
dict1= {'Film Name':title1, 'Director':dir1, 'Cast':cast1, 'Country':con1}
HF15 = pd.DataFrame(dict1)
HF15
```
