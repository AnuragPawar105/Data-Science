```python
import sys
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://store.steampowered.com/tags/en/Action/'

res=requests.get(url)

res.status_code
```

```python
soup=BeautifulSoup(res.content,'html.parser')
soup
```

```python
# TASK 1
games=soup.find_all('div',class_='tab_item_name')

gnames=[]
for i in games[1:-1]:
    gnames.append(i.text)
gnames
```

```python
gnames=pd.DataFrame(gnames,columns=['Trending Games'])
gnames
```

```python
price=soup.find_all('div',class_='discount_final_price')

gprices=[]
for i in price:
    gprices.append(i.text)
gprices
```

```python
gprices=pd.DataFrame(gprices,columns=['Prices'])
gprices
```

```python
# TASK 2
df=pd.concat([gnames,gprices],axis=1)
df
```
