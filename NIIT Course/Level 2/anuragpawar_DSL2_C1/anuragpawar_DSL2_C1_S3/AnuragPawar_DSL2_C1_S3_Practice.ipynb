```python
import requests
from bs4 import BeautifulSoup
```

```python
url= "https://en.wikipedia.org/wiki/Leonardo_DiCaprio_filmography"
response= requests.get(url).text
soup= BeautifulSoup(response, "lxml")
print(soup.prettify)
```

```python
# TASK 1
table= soup.find('table')
table
```

```python
films=[]
for row in table.find_all('tr')[1:]:
    columns = row.find_all('td')[0:2]
    films.append([column.text for column in columns])
films
```

```python
year = []
for row in table.find_all("tr")[1:]:
    temp= []
    for i in row.find_all("th"):
        k=i.text.replace("\n",'')
        temp.append(k)
    year.append(temp)
for i in range(0,len(year)):
    if year[i]==[]:
        year[i]= year[i-1]
year
```

```python
wyear= []
for i,j in list(enumerate(zip(year,films))):
    wyear.append(j[0]+j[1])
wyear
```

```python
import pandas as pd
df = pd.DataFrame(wyear, columns=['Year','Film Name','Role'])
df['Film Name'] = df['Film Name'].str.replace('\n','')
df['Role'] = df['Role'].str.replace('\n','')
df
```

```python
f= open('tab1.html','w')
table1= df.to_html()
f.write(table1)
f.close()
```
